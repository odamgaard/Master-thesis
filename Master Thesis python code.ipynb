### 1. Quantities and prices for Cournot-Cournot

#### 1.1 Total quantities produced

```python
# Define beta
beta = np.linspace(0, 1, 500)

# Define quantities
S1 = 4 / (3 * (2 + beta))
S2 = 2 / (4 * (1 + beta))
S3 = 2 / (2 + beta)
S4 = 1 / (2 * (2 + beta)) + (4 + beta) / (4 * (2 + beta))
S5 = 2 / (2 * (2 + beta))
S6 = 2 / (3 * (1 + beta))

# Create plot
plt.style.use('ggplot')
plt.figure(figsize=(8, 6))

# Plot each market configuration
plt.plot(beta, S1, label="S1", color='#700000', linewidth=2, marker='o', markersize=6, markevery=40)  
plt.plot(beta, S2, label="S2", color='#6c799d', linewidth=2, marker='s', markersize=6, markevery=40)  
plt.plot(beta, S3, label="S3", color='#a93226', linewidth=2, marker='D', markersize=6, markevery=40)  
plt.plot(beta, S4, label="S4", color='#000000', linewidth=2, marker='x', markersize=6, markevery=40)  
plt.plot(beta, S5, label="S5", color='#5d6d7e', linewidth=2, marker='^', markersize=6, markevery=40)  
plt.plot(beta, S6, label="S6", color='#154360', linewidth=2, marker='v', markersize=6, markevery=40)  
plt.gca().set_facecolor('#f7f7f7') 
plt.grid(True, linestyle='-', color='white', alpha=0.7) 

# Labels and legend
plt.xlabel(r'$\beta$', fontsize=16, weight='bold')
plt.ylabel('Total quantities', fontsize=14)  
plt.legend(loc='upper right', fontsize=10, fancybox=True, framealpha=1, shadow=True, borderpad=0.5, facecolor='white', markerscale=0.8)
plt.xlim(0, 1)
plt.ylim(0.1, 1)  
plt.xticks(np.linspace(0.2, 1, 5), fontsize=12)
plt.yticks(np.linspace(0.2, 1, 5), fontsize=12)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_color('#b0b0b0')
plt.gca().spines['bottom'].set_color('#b0b0b0')

# Show plot
plt.tight_layout()
plt.show()
```
